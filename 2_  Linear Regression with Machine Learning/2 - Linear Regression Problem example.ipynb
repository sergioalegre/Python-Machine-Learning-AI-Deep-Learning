```python
#https://www.eduonix.com/dashboard/complete-guide-to-machine-learning-using-python/lecture/16108022

import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model

swiss = datasets.make_swiss_roll() # este dataset tiene 3 features, vamos a estudiar la relaci√≥n entre las 2 primeras
# print(swiss)

x = swiss[0]
y = swiss[1]
# print(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

lm = linear_model.LinearRegression(normalize=True)

model = lm.fit(x_train,y_train)
predictions = lm.predict(x_test)


fig, ax = plt.subplots()
plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, predictions,'bo')
ax.grid()
plt.show()
```
